```python
import pandas
import numpy
```

```python
pcalls = '/Users/mf019/bioinformatics/longread_pangenome/plasmid_id/output/best_matches_v9_1kb.tsv'
tats = 'longread_report_TASmania.tsv'
```

```python
tats_df = pandas.read_csv(tats, sep='\t')
pcalls_df = pandas.read_csv(pcalls, sep='\t')
```

```python
tats_df
```

```python
pcalls_df['contig']
```

```python
pcalls_df = pcalls_df.rename(columns={'name': 'genome_id'})
```

```python
def process_best_hit(x):
    if pandas.notna(x) and '|' in x:
        return x.split('|')[-1]
    return x

pcalls_df['best_hit'] = pcalls_df['best_hit'].apply(process_best_hit)
```

```python
merged_df = pandas.merge(tats_df, pcalls_df[['genome_id', 'contig', 'best_hit']],
                     on=['genome_id', 'contig'],
                     how='left')
```

```python
# Update 'contig' column with 'best_hit' values where 'best_hit' is not NaN
merged_df['contig_id'] = numpy.where(merged_df['best_hit'].notna(), merged_df['best_hit'], merged_df['contig_id'])
```

```python
merged_df = merged_df.drop(columns=['best_hit'])
```

```python
merged_df.to_csv('longread_report_TASmania_fixed.tsv', sep='\t', index=False)
```
