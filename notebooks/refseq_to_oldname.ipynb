```python
import gffutils
import sys
import os
import pickle
import pprint
from collections import defaultdict
```

```python
gff_file = '/Users/mf019/bioinformatics/longread_GWAS/ref/GCF_000008685.2_ASM868v2_genomic.gff'
plasmid_dict_file = '/Users/mf019/bioinformatics/longread_GWAS/plasmid_id/blast/blast_parsing_dict.pickle'
```

```python
with open(gff_file) as f:
    gff = f.read()
```

```python
plasmid_dict = pickle.load(open(f'{plasmid_dict_file}', 'rb'))
pprint.pprint(plasmid_dict)
```

```python
genes_dict = defaultdict(dict)
for line in gff.split('\n'):
    if line.startswith('#'):
        continue
    if line.startswith('NC_'):
        #print(line.split('\t'))#[8].split(';'))
        if 'old_locus_tag' in line:
            replicon = line.split('\t')[0]
            annotations = line.split('\t')[8].split(';')
            #print(annotations)
            geneid = annotations[0].split('=')[-1].replace('gene-', '')
            genename = annotations[1].split('=')[-1]
            old_locus_tag = [x for x in annotations if 'old_locus_tag' in x][0].split('=')[-1]
            #print(f'Replicon={replicon}\tID={geneid}\tName={genename}\tlocus_tag={old_locus_tag}')
            genes_dict[geneid] = {'replicon': replicon, 'name': genename, 'locus_tag': old_locus_tag}
        else:
            print(line.split('\t'))
```

```python
pprint.pprint(genes_dict)
```

```python
pickle.dump(genes_dict, open('genes_dict.pickle', 'wb'))
```
