```python
import shutil
import glob
```

```python
path_to_asms = '../assemblies/paired_assemblies/'
path_to_paired_only = '/Users/mf019/bioinformatics/longread_GWAS/assemblies/paired_assemblies/paired_only/'
```

```python
list_of_files = glob.glob(f'{path_to_asms}/contigs/*.fasta')
list_of_gffs = glob.glob(f'{path_to_asms}/annotation/gff/*.gff')
```

```python
print(len(list_of_files))
```

```python
list_of_asms = []
for file in list_of_files:
    isolate_name = file.split('/')[-1].split('.')[0]
    print(isolate_name)
    list_of_asms.append(isolate_name)
```

```python
illumina_assemblies = [isolate for isolate in list_of_files if not isolate.split('/')[-1].split('.')[0].endswith(('H', 'P'))]
print(illumina_assemblies)
```

```python
paired_asms = []
for file in illumina_assemblies:
    isolate_name = file.split('/')[-1].split('.')[0]
    if isolate_name + 'H' or isolate_name + 'P' in list_of_asms:
        print(f'{isolate_name} has paired assembly')
        paired_asms.append(isolate_name)
```

```python
for asm in paired_asms:
    shutil.copy(f'{path_to_asms}/contigs/{asm}.fasta', f'{path_to_paired_only}/shortread/contigs/{asm}.fasta')
    shutil.copy(f'{path_to_asms}/annotation/gff/{asm}.gff3', f'{path_to_paired_only}/shortread/annotation/{asm}.gff3')
    if asm + 'H' in list_of_asms:
        print(f'{asm}H is in list of assemblies')
        shutil.copy(f'{path_to_asms}/contigs/{asm}H.fasta', f'{path_to_paired_only}/longread/contigs/{asm}H.fasta')
        shutil.copy(f'{path_to_asms}/annotation/gff/{asm}H.gff3', f'{path_to_paired_only}/longread/annotation/{asm}H.gff3')
        print(f'{asm}H copied to longread directory')
    else:
        if asm + 'P' in list_of_asms:
            print(f'{asm}P is in list of assemblies')
            shutil.copy(f'{path_to_asms}/contigs/{asm}P.fasta', f'{path_to_paired_only}/longread/contigs/{asm}P.fasta')
            shutil.copy(f'{path_to_asms}/annotation/gff/{asm}P.gff3', f'{path_to_paired_only}/longread/annotation/{asm}P.gff3')
            print(f'{asm}P copied to longread directory')
```

```python
print('number of shortread asms:',len(glob.glob(f'{path_to_paired_only}/shortread/contigs/*.fasta')))
print('number of longread asms:',len(glob.glob(f'{path_to_paired_only}/longread/contigs/*.fasta')))

print('number of shortread gffs:',len(glob.glob(f'{path_to_paired_only}/shortread/annotation/*.gff3')))
print('number of longread gffs:',len(glob.glob(f'{path_to_paired_only}/longread/annotation/*.gff3')))
```
