```python
import os
from collections import defaultdict
from Bio import SeqIO
```

```python
def parse_genbank(path):
    records = defaultdict()
    for record in SeqIO.parse(path, 'genbank'):
        records[record.id] = record
    return records
```

```python
def split_genbanks(records, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    for record in records:
        active_record = records[record]
        rec_id = active_record.name
        output_file = f'{output_dir}/{rec_id}.gbff'
        with open(output_file, 'w') as outfile:
            SeqIO.write(active_record, outfile, "genbank")
```

```python
bb_ss_genbank = 'BbSs.gbff'
plasmids = parse_genbank(bb_ss_genbank)
split_dir = 'single_gbs'
split_genbanks(plasmids, split_dir)
```
