```python
import glob
import shutil
import pathlib
import os
from collections import defaultdict
```

```python
list_o_results = glob.glob('/mnt/disks/big-results/longread_pangenome_alignments/asms_all_v_all/homology/*')
```

```python
print(len(list_o_results))
```

```python
new_tld = os.path.join('/mnt/disks/big-results/longread_pangenome_alignments/asms_all_v_all/', 'fixed')
#os.makedirs(new_tld)
```

```python
paths_dict = defaultdict(dict)
```

```python
for result in list_o_results:
    basename = os.path.basename(result)
    comp1 = basename.split('_vs_')[0]
    comp2 = basename.split('_vs_')[1]
    asm1 = '_'.join(comp1.split('_')[0:-1])
    contig1 = comp1.split('_')[-1]
    asm2 = '_'.join(comp2.split('_')[0:-1])
    contig2 = comp2.split('_')[-1]
    if contig1 not in paths_dict[asm1]:
        paths_dict[asm1].update({contig1: []})
    paths_dict[asm1][contig1].append(basename)
    #print(f'path:{result}\npath2:{new_output_dir}\nbasename:{basename}\ncomp1:{comp1}\ncomp2:{comp2}\nasm1:{asm1} contig1:{contig1}\nasm2:{asm2} contig2:{contig2}\n')
```

```python
print(len(paths_dict))
```

```python
for key1, items1 in paths_dict.items():
    for key2, items2 in items1.items():
        print(f'{key1}: {key2}: {len(items2)}')
```
