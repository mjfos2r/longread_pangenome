```python
test_file = 'B331PP.long_stats.txt'
```

```python
import re
```

```python
with open(test_file, 'r') as in_f:
    feature_types = ['region', 'mrna', 'ncrna', 'rrna', 'tmrna', 'trna']
    parsed_data = {feat_type: {} for feat_type in feature_types}
    current_section = None
    
    for line in in_f.readlines():
        if line.startswith('----'):
            section_match = re.search(r'-+ ([a-z]+) -+', line)
            if section_match:
                current_section = section_match.group(1)
            continue
        
        # Skip empty lines
        if not line.strip():
            continue
            
        if current_section:
            # Split only on multiple spaces
            parts = re.split(r'\s{2,}', line.strip())
            if len(parts) == 2:
                key = parts[0]
                value = parts[1]
                try:
                    if value != 'na':
                        value = float(value)
                    if current_section in feature_types:
                        parsed_data[current_section][key] = value
                except ValueError:
                    continue
```
