```python
shortread_genomes = [
    'GCF_000008685.2_ASM868v2_genomic',
    'GCF_000021405.1_ASM2140v1_genomic',
    'GCF_000166635.2_ASM16663v1_genomic',
    'GCF_000166655.2_ASM16665v1_genomic',
    'GCF_000171735.2_ASM17173v2_genomic',
    'GCF_000172295.2_ASM17229v2_genomic',
    'GCF_000172315.2_ASM17231v2_genomic',
    'GCF_000172335.2_ASM17233v2_genomic',
    'GCF_000181555.2_ASM18155v2_genomic',
    'GCF_000181575.2_ASM18157v2_genomic',
    'GCF_000181715.2_ASM18171v2_genomic',
    'GCF_000181855.2_ASM18185v2_genomic'
]
dropped_genomes = [
    'GCA_030439625.1_ASM3043962v1_genomic',
    'GCA_032595535.1_ASM3259553v1_genomic',
    'GCF_000172295.2_ASM17229v2_genomic',
    'GCF_000181715.2_ASM18171v2_genomic',
    'URI45H',
    'GCA_030445025.1_ASM3044502v1_genomic',
    'GCF_000181575.2_ASM18157v2_genomic',
    'GCA_030445005.1_ASM3044500v1_genomic',
    'GCA_030439605.1_ASM3043960v1_genomic',
    'GCA_032595755.1_ASM3259575v1_genomic',
    'GCF_000008685.2_ASM868v2_genomic',
    'UCT95H',
    'GCA_032595515.1_ASM3259551v1_genomic',
    'GCA_038801875.1_ASM3880187v1_genomic',
    'GCA_032595375.1_ASM3259537v1_genomic',
    'GCA_030439665.1_ASM3043966v1_genomic',
    'GCF_000172315.2_ASM17231v2_genomic',
    'GCF_000181855.2_ASM18185v2_genomic',
    'GCA_030445065.1_ASM3044506v1_genomic',
    'GCA_030444985.1_ASM3044498v1_genomic',
    'GCF_000172335.2_ASM17233v2_genomic',
    'GCA_030440345.1_ASM3044034v1_genomic',
    'GCF_000021405.1_ASM2140v1_genomic',
    'GCF_000181555.2_ASM18155v2_genomic',
    'GCF_000171735.2_ASM17173v2_genomic',
]
duplicated_genomes = [
    'GCF_002442595.2_ASM244259v2_genomic',
    'GCF_019464575.1_ASM1946457v1_genomic',
    'GCF_024662155.1_ASM2466215v1_genomic',
    'GCF_024662175.1_ASM2466217v1_genomic',
]

def check_duplicates(directory):
    keep_count = 0
    remove_count = 0
    files = os.listdir(directory)
    for file in files:
        if file in dropped_genomes:
            remove_count += 1
            print(f"Removed: {file}")
        elif file.startswith('GCA_'):
            gcf_version = file.replace('GCA_', 'GCF_')
            if gcf_version in files:
                #os.remove(file)
                remove_count += 1
                print(f"Removed: {file}")
            else:
                print(f"Kept: {file}")
                keep_count += 1
        
        elif file in duplicated_genomes:
            remove_count += 1
            print(f"Duplicate: {file}")
                
        else:
            print(f"Kept: {file}")
            keep_count += 1
    print(keep_count)
    print(remove_count)
```
