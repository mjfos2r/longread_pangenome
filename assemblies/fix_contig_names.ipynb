```python
import os
import glob
from collections import defaultdict
from pathlib import Path
import re
import shutil
```

```python
asm_dir = 'jel'
log_files = glob.glob(f'{asm_dir}/*/*.log')
```

```python
pattern = r'id=([\w_]+),\s*orig-id=([\w\d]+)'
header = 'original sequence id\tnew sequence id\ttype\ttopology\tname\n'

for file in log_files:
    sample_name = Path(file).stem
    naming_table = Path(file).with_suffix('.contig_names.tsv')
    print(sample_name)
    with open(file, 'r') as f:
        with open(naming_table, 'w') as o:
            o.write(header)
            lines = f.readlines()
            for line in lines:
                line = line.strip()
                if '- INFO - UTILS - qc: revised sequence:' in line:
                    match = re.search(pattern, line)
                    if match:
                        key = match.group(1)    # contig_1
                        value = match.group(2)  # contig000001
                        o.write(f'{key}\t{value}\t-\t-\t-\n')
```
