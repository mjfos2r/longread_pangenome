```python
kept_list = '/home/mf019/longread_pangenome/expanded_dataset_analysis/misc/kept_genomes_v4.txt'
with open(kept_list) as f:
    lines = [line.strip() for line in f]
```

```python
for idx, line in enumerate(lines):
    if line.startswith('GCF'):
        lines[idx] = '_'.join(line.split('_')[0:2])
        
```

```python
count = 0
kept_rs = []
for line in lines:
    if line.startswith('GCF'):
        print(line.split('.')[0])
        count += 1
        kept_rs.append(line)
print(count)
```
