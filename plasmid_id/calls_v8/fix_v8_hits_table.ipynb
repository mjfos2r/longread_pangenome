```python
import pandas
```

```python
v8_table = pandas.read_csv('best_matches_v8_1kb.tsv', sep='\t')
```

```python
v8_table
```

```python
v8_table[['wp_alignment_id', 'wp_plasmid_name', 'pf_alignment_id', 'pf_plasmid_name']]
v8_table['pf_alignment_id'] = v8_table['pf_alignment_id'].fillna(0)
v8_table['pf_plasmid_name'] = v8_table['pf_plasmid_name'].fillna(0)
```

```python
for index, row in v8_table.iterrows():
    #print(f"Row {index}:")
    #print(f"  wp_alignment_id: {row['wp_alignment_id']}")
    #print(f"  wp_plasmid_name: {row['wp_plasmid_name']}")
    #print(f"  pf_alignment_id: {row['pf_alignment_id']}")
    #print(f"  pf_plasmid_name: {row['pf_plasmid_name']}")
    if row['pf_alignment_id'] != 0:
        #print(row['pf_alignment_id'])
        if "ParA" in row['pf_alignment_id']:
            fixed_pf_name = row['pf_alignment_id'].split('_')[-3]
            #print(fixed_pf_name)
        else:
            fixed_pf_name = row['pf_alignment_id'].split('_')[-1]
            #print(fixed_pf_name)
        v8_table.at[index, 'pf_plasmid_name'] = fixed_pf_name
        v8_table.at[index, 'best_hit'] = fixed_pf_name
    else:
        v8_table.at[index, 'best_hit'] = row['wp_plasmid_name']
```

```python
columns = ['name', 'method', 'contig', 'contig_len', 'wp_alignment_id', 'pf_alignment_id', 'wp_plasmid_name', 'pf_plasmid_name', 'best_hit']
v9_subset = v8_table[columns]
```

```python
v9_subset.to_csv('best_matches_v9_1kb.tsv', sep='\t', index=False)
```
