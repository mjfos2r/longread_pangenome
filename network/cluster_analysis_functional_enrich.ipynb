```python
import pandas
import gffutils
import pickle
db_file = '../ref/B31_gffdb.db'
sets_pickle = 'full_network_kpartite_sets.pkl'
cluster_annotations_pickle = '../notebooks/metadb/clustered_proteins_db_results_v1.pkl'
```

```python
with open(sets_pickle, 'rb') as infile:
    kp_sets = pickle.load(infile)

with open(cluster_annotations_pickle, 'rb') as infile:
    group_annos = pickle.load(infile)
```

```python
group_annos['Lipoprotein']['locus_tag']
```

```python
from collections import defaultdict


cols = ['locus_tag', 'product', 'replicon_name', 'PFAM', 'GO', 'SO', 'COG', 'EC']
sets = ['locus_tag', 'product', 'replicon_name', 'PFAM', 'GO', 'SO', 'COG', 'EC']

group_dict = defaultdict(dict)
for item in group_annos:
    group_dict[item] = {'locus_tag': set(), 'product': set(), 'replicon_name': set(), 'PFAM': set(),
                        'GO': set(), 'SO': set(), 'COG': set(), 'EC': set()}
    sub = group_annos[item][cols]
    for annotation in sets:
        items = group_annos[item][annotation].to_list()
        for i in items:
            group_dict[item][annotation].add(i)
```

```python
def process_entry(entry):
    return {k: (list(v) if v else None) for k, v in entry.items()}
```

```python
kp_sets_dfs = []

for i in kp_sets:
    rows = []
    for group in kp_sets[i]:
        rows.append(group_dict[group])
    print(rows)
    collapsed_row = [process_entry(entry) for entry in rows]
    collapsed_row.append({'group_id': group})
    kp_sets_dfs.append(pandas.DataFrame(collapsed_row))
```
